```python
import json
first_time_1 = "/Users/gongshukai/Desktop/result_uspto_request backup/15090-18089/uspto_multiple_step/uspto_multiple_step1.json"
second_time_1 = "/Users/gongshukai/Desktop/result_uspto_request backup/15090-18089/uspto_multiple_step/uspto_multiple_step2.json"

def count_file_steps(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        total_data = [json.loads(line) for line in lines]
        # Create a dictionary to store the length of reaction. The key is the length of reaction, the value is the number of reactions.
        reaction_length = {}
        for reaction in total_data:
            length = len(reaction['response'].split('\n'))
            if length in reaction_length:
                reaction_length[length] += 1
            else:
                reaction_length[length] = 1
    return reaction_length


def statistic_dataset(first_time_1, second_time_1):
    first_time_valid_reactions = 0
    second_time_valid_reactions = 0
    with open(first_time_1, 'r') as first_time:
        lines = first_time.readlines()
        first_time_valid_reactions += len(lines)
    with open(second_time_1, 'r') as second_time:
        lines = second_time.readlines()
        second_time_valid_reactions += len(lines)
    dict1 = count_file_steps(first_time_1)
    dict2 = count_file_steps(second_time_1)
    print("First time API calls has " + str(first_time_valid_reactions) + " valid reactions.")
    print(dict1)
    print([dict1[key] / first_time_valid_reactions for key in dict1])
    print("Second time API calls has " + str(second_time_valid_reactions) + " valid reactions.")
    print(dict2)
    print([dict2[key] / second_time_valid_reactions for key in dict2])
    

statistic_dataset(first_time_1, second_time_1)
```

```python
# 现在，我有一个文件夹result_uspto_request backup
# 里面有许多文件夹（例如0-89, 90-1589, 1590-3089)文件夹的数字没有规律
# 每个文件夹里面有个文件夹uspto_multiple_step，里面有个文件uspto_multiple_step.json
# 写一个函数，遍历每个result_uspto_request backup的子文件夹，把每个子文件夹所有uspto_multiple_step.json的文件目录合并到一个list中，返回这个list

import os

def find_uspto_json_files(root_dir, folder_name):
    json_files_list = []
    subfolders = []

    # 收集所有一级子文件夹并解析其名称为数字范围
    for entry in os.listdir(root_dir):
        subfolder_path = os.path.join(root_dir, entry)
        
        # 确保这是一个目录
        if os.path.isdir(subfolder_path):
            # 尝试从文件夹名称中提取数字范围
            try:
                # 分割文件夹名，取其开始数字
                range_start = int(entry.split('-')[0])
                # 将起始数字和文件夹路径作为元组存储
                subfolders.append((range_start, subfolder_path))
            except ValueError:
                # 如果文件夹名不符合预期格式，跳过
                continue
    
    # 根据数字范围的起始值排序子文件夹
    subfolders.sort()

    # 遍历排序后的子文件夹
    for _, subfolder_path in subfolders:
        # 构建uspto_multiple_step文件夹的路径
        uspto_folder_path = os.path.join(subfolder_path, 'uspto_multiple_step')
        
        # 检查uspto_multiple_step文件夹是否存在
        if os.path.exists(uspto_folder_path) and os.path.isdir(uspto_folder_path):
            # 构建json文件的路径
            json_file_path = os.path.join(uspto_folder_path, 'uspto_multiple_step.json')
            
            # 如果文件存在，则添加到列表
            if os.path.exists(json_file_path):
                json_files_list.append(json_file_path)
    
    return json_files_list

rootdir = '/Users/gongshukai/Desktop/result_uspto_request backup'
folder_name = 'uspto_multiple_step'
write_path = '/Users/gongshukai/Desktop/result_uspto_request backup/valid_reactions/uspto_multiple_step.json'
multiple_step_paths = find_uspto_json_files(rootdir, folder_name)
merge_files(multiple_step_paths, write_path)
statistic_dataset(write_path, write_path)
```
